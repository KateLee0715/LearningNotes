```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as scio
import math
import matplotlib
import matplotlib.image as mpimg
```

```python
data = scio.loadmat("ex7data1.mat")
X = data['X']
```

```python
plt.plot(X[:, 0], X[:, 1], 'bo', markerfacecolor = 'w')
plt.axis([0.5, 6.5, 2.0, 8.0])
```

```python
def pca(X):
    m, n = X.shape
    sigma = np.dot(X.T, X) / m
    U, S, V = np.linalg.svd(sigma)
    return U, S
```

```python
def featureNormalize(X):
    mu = np.mean(X, axis = 0)
    sigma = np.std(X - mu, axis = 0)
    X_norm = (X - mu) / sigma
    return X_norm, mu, sigma
```

```python
X_norm, mu, sigma = featureNormalize(X)
U, S = pca(X_norm)
U[:,0]
```

```python
p1 = mu
p2 = mu + 1.5 * S[0] * U[:,0].T
p3 = mu + 1.5 * S[1] * U[:,1].T
plt.plot(X[:, 0], X[:, 1], 'bo', markerfacecolor = 'w')
plt.axis([0.5, 6.5, 2.0, 8.0])
drawLine(p1, p2, 'k-', 2)
drawLine(p1, p3, 'k-', 2)
```

```python
def projectData(X, U, K):
    U_reduce = U[:, 0:K]
    X_pro = np.dot(X, U_reduce)
    return X_pro
```

```python
X_pro = projectData(X_norm, U, 1)
X_pro[0, :]
```

```python
def recoverData(Z, U, K):
    U_reduce = U[:, 0 : K]
    X = np.dot(Z, U_reduce.T)
    return X
```

```python
X_rec = recoverData(X_pro, U, 1)
X_rec[0, :]
```

```python
def drawLine(p1, p2, c, linewidth):
    plt.plot([p1[0], p2[0]], [p1[1], p2[1]], c, linewidth = linewidth)
```

```python
plt.plot(X_norm[:, 0], X_norm[:, 1], 'bo', markerfacecolor = 'w')
plt.plot(X_rec[:, 0], X_rec[:, 1], 'ro', markerfacecolor = 'w')
plt.axis([-4, 3, -4, 3])
for i in range(X_norm.shape[0]):
    drawLine(X_norm[i, :], X_rec[i, :], 'k--', 1)
```

```python
def displayData(X, example_width):
    m, n = X.shape
    example_width = int(example_width)
    example_height = int(n / example_width)
    display_rows = int(np.floor(math.sqrt(m)))
    display_cols = int(np.ceil(m / display_rows))
    h, display_arr = plt.subplots(nrows = display_rows, ncols = display_cols, sharex = True, sharey = True, figsize = (8, 8))
    plt.subplots_adjust(wspace = 0.05, hspace = 0.05)
    for i in range(display_rows):
        for j in range(display_cols):
            display_arr[i, j].matshow(np.array(X[i * display_cols + j]).reshape(example_height, example_width).T, cmap = matplotlib.cm.gray)
            plt.xticks(np.array([]))
            plt.yticks(np.array([]))
    return h, display_arr
```

```python
def kMeansInitCentroids(X, K):
    m = X.shape[0]
    random_indices = np.random.choice(np.arange(m), K)
    centroids = X[random_indices, :]
    return centroids.T
```

```python
def findClosestCentroids(X, centroids):
    m, n = X.shape
    idx = np.zeros(m)
    for i in range(X.shape[0]):
        tmp = X[i, :] - centroids[:,0].reshape(1, n)
        minJ = np.dot(tmp, tmp.T)
        for j in range(1, centroids.shape[1]):
            tmp = X[i, :] - centroids[:,j].reshape(1, n)
            tmpJ = np.dot(tmp, tmp.T)
            if tmpJ < minJ:
                minJ = tmpJ
                idx[i] = j
    return idx
```

```python
def computeCentroids(X, idx, K):
    n = X.shape[1]
    centroids = np.zeros((n, K))
    for i in range(K):
        centroids[:, i] = np.dot(X.T, idx == i) / np.sum(idx == i)
    return centroids
```

```python
def runKMeans(X, init_centroids, max_iters):
    m = X.shape[0]
    idx = np.zeros(m)
    K = init_centroids.shape[1]
    centroids = init_centroids
    for i in range(max_iters):
        idx = findClosestCentroids(X, centroids)
        centroids = computeCentroids(X, idx, K)
    return centroids, idx
```

```python
bird = mpimg.imread("bird_small.png")
m, n, x = bird.shape
X_bird = bird.reshape(m * n, x)
K = 16
max_iters = 10
init_centroids = kMeansInitCentroids(X_bird, K)
centroids, idx = runKMeans(X_bird, init_centroids, max_iters)
```

```python
random_indices = np.random.choice(np.arange(m * n), 1000)
```
