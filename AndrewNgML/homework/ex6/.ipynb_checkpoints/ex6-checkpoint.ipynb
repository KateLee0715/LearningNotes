```python
from libsvm.svmutil import *
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as scio
```

```python
data1 = scio.loadmat("ex6data1.mat")
X1 = data1['X']
y1 = data1['y']
```

```python
def plotData(X, y):
    pos = np.where(y == 1)
    neg = np.where(y == 0)
    plt.plot(X[pos, 0], X[pos, 1], 'k+', linewidth = 2, markersize = 8)
    plt.plot(X[neg, 0], X[neg, 1], 'ko', markerfacecolor = 'y', markersize = 8)
```

```python
plotData(X1, y1)
```

```python
def svmTrain(X, y, options):
    m = svm_train(y.reshape(-1), X, options)
    sv_x1 = []
    sv_y1 = []
    sv_temp = []
    # 找支持向量SV的坐标
    sv1 = m.get_SV()
    # 找支持向量SV在训练集中的序号，从1开始，注意与列表序号区别，需要减一，y2[sv_idx[i]-1]
    sv_idx1 = m.get_sv_indices()

    for i in range(len(sv_idx1)):
        sv_y1.append(int(y[sv_idx1[i] - 1]))  # 通过序号找到标签
        for j in range(len(sv1[0])):
            sv_temp.append(sv1[i][j + 1])  # 特征序号从1，开始，故需j + 1
        # print(x1[sv_idx1[i]-1])
        sv_x1.append(sv_temp)
        sv_temp = []
    sv_x1 = np.array(sv_x1).reshape(i + 1, j + 1)
    # print(sv1)
    # print(sv_y1)

    # 系数alpha*y
    alpha1 = m.get_sv_coef()
    
    w = 0.0
    for i in range(len(alpha1)):
        # w = w + abs(alpha1[i][0]) * sv_y1[i] * sv_x1[i]  # 求w的公式
        w = w + alpha1[i][0] * sv_x1[i] 
    
    b = -m.rho[0]
    return w, b
```

```python
def visualizeBoundaryLinear(X, y, options):
    w, b = svmTrain(X, y, options)
    xp = np.linspace(np.amin(X[:,0]), np.amax(X[:,0]), 100)
    yp = - w[0] / w[1] * xp - b / w[1]
    plotData(X1, y1)
    plt.plot(xp, yp, 'b-', linewidth = 2)
```

```python
visualizeBoundaryLinear(X1, y1, '-c 1 -t 0')
```

```python
visualizeBoundaryLinear(X1, y1, '-c 100 -t 0')
```

```python
def gaussianKernel(X1, X2, sigma):
    x_minus = X1 - X2
    k = np.exp(- np.dot(x_minus.T, x_minus) / (2 * sigma * sigma))
    return k
```

```python
x1 = np.array([1, 2, 1])
x2 = np.array([0, 4, -1])
sigma = 2
sim = gaussianKernel(x1, x2, sigma)
print(sim)
```

```python
data2 = scio.loadmat("ex6data2.mat")
X2 = data2['X']
y2 = data2['y']
```

```python
plotData(X2, y2)
```

```python
def visualizeBoundary(X, y, options):
    m = svm_train(y.reshape(-1), X, options)
    x1 = np.linspace(np.amin(X[:,0]), np.amax(X[:,0]), 100).T
    x2 = np.linspace(np.amin(X[:,1]), np.amax(X[:,1]), 100).T
    x1, x2 = np.meshgrid(x1, x2)
    vals = np.zeros((x1.shape[0], x1.shape[0]))
    yy = np.zeros(x1.shape[0])
    for i in range(x1.shape[1]):
        this_x = np.vstack((x1[:,i], x2[:,i]))
        a, b, val = svm_predict(yy, this_x.T, m)
        vals[:,i] = np.array(a).reshape(-1)
    plotData(X, y)
    plt.contour(x1, x2, vals,[0.5], colors = 'blue')
    return
```

```python
visualizeBoundary(X2, y2, '-c 1 -t 2 -g 50')
```

```python
data3 = scio.loadmat("ex6data3.mat")
X3 = data3['X']
y3 = data3['y']
Xval = data3['Xval']
yval = data3['yval']
```

```python
plotData(X3, y3)
```

```python
def dataset3Params(X, y, Xval, yval):
    tmpC = 0
    tmpSigma = 0
    tmpAcc = 0
    steps = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30]
    for i in range(len(steps)):
        for j in range(len(steps)):
            options = '-t 2 -c ' + str(steps[i]) + ' -g ' + str(1 / (2 * steps[j] * steps[j]))
            m = svm_train(y.reshape(-1), X, options)
            p_label, p_acc, p_val = svm_predict(yval.reshape(-1), Xval, m)
            if p_acc[0] > tmpAcc:
                tmpC = steps[i]
                tmpSigma = steps[j]
                tmpAcc = p_acc[0]
    return tmpC, tmpSigma
```

```python
C, sigma = dataset3Params(X3, y3, Xval, yval)
```

```python
print(C)
print(sigma)
```

```python
options = '-t 2 -c ' + str(C) + ' -g ' + str(1 / (2 * sigma * sigma))
visualizeBoundary(X3, y3, options)
```
